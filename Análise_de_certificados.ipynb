# **Visualização de Certificados**

```python
import pandas as pd
import os
from google.colab import drive
from IPython.display import HTML, display

# Montar o Google Drive
drive.mount('/content/drive')

# Caminho para salvar o arquivo CSV no Google Drive
drive_path = '/content/drive/MyDrive/Certificados'
if not os.path.exists(drive_path):
    os.makedirs(drive_path)
path = os.path.join(drive_path, 'certificados.csv')

# Verifica se o arquivo já existe
if os.path.exists(path):
    df_certificados = pd.read_csv(path)
else:
    df_certificados = pd.DataFrame(columns=["ID", "Nome", "Curso", "Data", "Carga Horária", "Instituição", "Link"])

def visualizar_previsualizacao():
    global df_certificados
    if df_certificados.empty:
        display(HTML("<p>Nenhum certificado encontrado.</p>"))
    else:
        # Ordenar por carga horária decrescente
        df_certificados = df_certificados.sort_values(by='Carga Horária', ascending=False)

        # Criando uma tabela HTML para exibição
        html_table = "<h3>Pré-visualização dos Certificados</h3>"
        html_table += "<table><tr><th>Curso</th><th>Data</th><th>Instituição</th><th>Link</th></tr>"

        for index, row in df_certificados.iterrows():
            html_table += "<tr>"
            html_table += f"<td>{row['Curso']}</td>"
            html_table += f"<td>{row['Data']}</td>"
            html_table += f"<td>{row['Instituição']}</td>"
            if pd.notna(row['Link']):
                html_table += f"<td><a href='{row['Link']}' target='_blank'>Acessar</a></td>"
            else:
                html_table += "<td>N/A</td>"
            html_table += "</tr>"

        html_table += "</table>"

        display(HTML(html_table))

# Mostrar a pré-visualização dos certificados
visualizar_previsualizacao()
```

```python
import pandas as pd
import gdown
from google.colab import drive
import matplotlib.pyplot as plt

# Bibliotecas de palavras-chave para categorização de cursos
biblioteca_ti = [
    "Técnico em Informática",
    "Redes",
    "Linux",
    "Cloud Computing",
    "AWS",
    "Segurança da Informação",
    "Bancos de Dados",
    "Sistemas Operacionais",
    "Virtualização",
    "Elétrica para Computadores",
    "ELÊTRICA PARA COMPUTADORES",
    "Eletrônica Analógica",
    "Implantação de Sistema PFSense",
    "Excel 2010",
    "CFTV",
    "Fundamentos de TI",
    "Principais Protocolos de Comunicação da Internet",
    "Exame: Fundamentos de Rede",
    "Exame: Fundamentos de Informática",
    "Introdução ao Sistema",
    "Introdução ao Conceito de Cloud",
    "Criando um Container de uma Aplicação Web",
    "ELETRÔNICA ANALÓGICA (REPARO) ESTABILIZADOR | NOBREAK| CIRCUITOS ELETRÔNICOS | MICROPROCESSADORES",
    "Telecomunicação e Redes"
]

biblioteca_dev = [
    "Python",
    "Java",
    "JavaScript",
    "Git",
    "Docker",
    "INTRODUÇÃO E LABORATÓRIO VIRTUAL",
    "INTRODUÇÃO À LGPD",
    "TRABALHANDO EM EQUIPES ÁGEIS",
    "O PODER DAS FUNÇÕES EM KOTLIN",
    "CONFIGURANDO SERVIDORES LINUX E INTRODUÇÃO AO DOCKER",
    "API REST",
    "DOMINANDO A LINGUAGEM DE PROGRAMAÇÃO KOTLIN",
    "Frameworks Web",
    "Programação Orientada a Objetos",
    "Testes Automatizados",
    "Desenvolvimento Ágil",
    "Desenvolvedor Front-End",
    "Bootcamp TQI Kotlin",
    "Algoritmo",
    "Avalie este Bootcamp",
    "Domindo a Linguagem de Programação Kotlin",
    "Inovação e Bioeconomia",
    "Metaverso: Verdades e Mitos",
    "INTRODUÇÃO A BANCO DE DADOS RELACIONAIS (SQL)",
    "Jornada Android",
    "Introdução à Internet das Coisas",
    "DevOps: Tudo o que você precisa saber",
    "Estruturas de Controle de Fluxo e Coleções em Kotlin",
    "Testes Unitários com JUnit",
    "Introdução ao Desenvolvimento Moderno de Software",
    "Introdução Prática à Linguagem de Programação Kotlin",
    "Tratamento de Exceções em Kotlin",
    "Orientação a Objetos e Tipos de Classes na Prática com Kotlin",
    "Conheça Tudo sobre a LGPD",
    "Fundamentos de BI: KPIs e Métricas",
    "MySQL",
    "Infraestrutura como Código",
    "Criando um Container de uma Aplicação Web",
    "Conhecendo o Kotlin e sua Documentação Oficial",
    "Contextualizando o Desenvolvimento Web com Spring Boot 3 e Kotlin",
    "Boas-Vindas ao Bootcamp TQI Kotlin",
    "Criando um Microsserviço de Upload de Imagens com o Amazon S3",
    "Algritmos de SVM",
    "Seja Protagonista neste Bootcamp",
    "Conheça as Oportunidades da DIO",
    "Banco de Dados: SQL e NoSQL",
    "INTRODUÇÃO A BANCO DE DADOS NOSQL",
    "Segurança da Informação"
]

biblioteca_ds = [
    "Introdução à Ciência de Dados",
    "Python para Ciência de Dados",
    "R para Análise de Dados",
    "Manipulação de Dados",
    "Visualização de Dados",
    "Machine Learning",
    "Big Data",
    "TRABALHANDO COM VISUAIS NO POWER BI",
    "Análise Estatística",
    "SQL para Ciência de Dados",
    "NLP",
    "Ética em Ciência de Dados",
    "Redes Neurais Artificiais",
    "KERAS",
    "DEEP LEARNING",
    "PYTHON DATA ANALYTICS",
    "COLETA E EXTRAÇÃO DE DADOS COM POWER BI",
    "TCE-ESTÁGIO-GRUNNER",
    "FUNDAMENTOS DE IA PARA DEVS",
    "Google Data Analytics",
    "Análise de Dados com Programação em R",
    "Analisar os Dados para Responder às Perguntas",
    "Compartilhar os Dados com a Arte da Visualização",
    "Processar os Dados para Limpá-los",
    "Preparar os Dados para Exploração",
    "Fundamentos: Dados, Dados, em Todos os Lugares",
    "Fazer Perguntas para Tomar Decisões com Base em Dados",
    "Projeto Final de Data Analytics do Google",
    "Fundamentos de BI: KPIs e Métricas",
    "Algoritmos Genéticos",
    "CERTIFICADO PROFISSIONAL DE ANÁLISE DE  DADOS DO GOOGLE",
    "Limpeza e Transformação de Dados com Power BI",
    "Criando Dashboard Interativos com Power BI",
    "Classificação de Problemas: Explorando Datasets",
    "Copiando Arquivos e Manipulando Processos",
    "Introdução à Inteligência Artificial",
    "INTELIGÊNCIA ARTIFICIAL APLICADA À VISÃO COMPUTACIONAL",
    "INTRODUÇÃO À PROGRAMAÇÃO E PENSAMENTO COMPUTACIONAL",
    "Desenvolvimento de Softwares"
]

biblioteca_outros = [
    "Curso de Satisfação do Cliente",
    "Gestão em Faturamento e Auditoria de Convênios Médicos",
    "Liderança: Líder x Chefe",
    "Administrando seu Dinheiro",
    "Empresário do Terceiro Milênio",
    "Logística para Faturamento de Convênios"
]

def montar_drive():
    """
    Monta o Google Drive para acessar o arquivo CSV.

    Returns:
        None
    """
    try:
        drive.mount('/content/drive', force_remount=True)
        print("Google Drive montado com sucesso.")
    except Exception as e:
        print(f"Erro ao montar Google Drive: {str(e)}")

def baixar_arquivo(file_id, caminho_arquivo):
    """
    Baixa o arquivo CSV do Google Drive usando o ID fornecido.

    Args:
        file_id (str): ID do arquivo no Google Drive.
        caminho_arquivo (str): Caminho onde o arquivo será salvo.

    Returns:
        None
    """
    try:
        url = f'https://drive.google.com/uc?id={file_id}'
        gdown.download(url, caminho_arquivo, quiet=False)
        print(f"Arquivo baixado com sucesso em {caminho_arquivo}.")
    except Exception as e:
        print(f"Erro ao baixar arquivo: {str(e)}")

def carregar_dataframe(caminho_arquivo):
    """
    Carrega o arquivo CSV em um DataFrame do Pandas.

    Args:
        caminho_arquivo (str): Caminho do arquivo CSV.

    Returns:
        pandas.DataFrame: DataFrame carregado com os dados do arquivo CSV.
    """
    try:
        return pd.read_csv(caminho_arquivo)
    except Exception as e:
        print(f"Erro ao carregar DataFrame: {str(e)}")
        return pd.DataFrame()

def categorizar_cursos(df):
    """
    Categoriza os cursos do DataFrame com base nas bibliotecas definidas.

    Args:
        df (pandas.DataFrame): DataFrame contendo os cursos a serem categorizados.

    Returns:
        pandas.DataFrame: DataFrame categorizado com as colunas 'Curso', 'Categoria', 'Instituição' e 'Carga Horária'.
    """
    categorias_encontradas = []

    for index, row in df.iterrows():
        curso = row['Curso']
        instituicao = row['Instituição']
        carga_horaria = row['Carga Horária']
        categoria_encontrada = None

        for biblioteca, nome_biblioteca in zip([biblioteca_ti, biblioteca_dev, biblioteca_ds, biblioteca_outros],
                                               ['Tecnologia da Informação e Infraestrutura',
                                                'Desenvolvimento e Programação',
                                                'Ciência de Dados e Analytics',
                                                'Outros']):
            for palavra_chave in biblioteca:
                if palavra_chave.lower() in curso.lower():
                    categoria_encontrada = nome_biblioteca
                    break
            if categoria_encontrada:
                break

        if not categoria_encontrada:
            categoria_encontrada = 'Não Especificada'

        categorias_encontradas.append([curso, categoria_encontrada, instituicao, carga_horaria])

    df_categorizado = pd.DataFrame(categorias_encontradas, columns=['Curso', 'Categoria', 'Instituição', 'Carga Horária'])
    return df_categorizado

def exibir_detalhes_cursos(categoria):
    """
    Exibe detalhes dos cursos de uma categoria específica.

    Args:
        categoria (str): Categoria dos cursos a serem exibidos.

    Returns:
        None
    """
    try:
        cursos_categoria = df_categorizado[df_categorizado['Categoria'] == categoria]
        cursos_categoria = cursos_categoria.sort_values(by='Carga Horária', ascending=False)

        if cursos_categoria.empty:
            print(f"\nNão foram encontrados cursos na categoria '{categoria}'.")
        else:
            print(f"\n### Categoria: {categoria} ###\n")
            for index, curso in cursos_categoria.iterrows():
                print(f"Curso: {curso['Curso']}")
                print(f"Instituição: {curso['Instituição']}")
                print(f"Carga Horária: {curso['Carga Horária']} horas")
                print("-" * 20)
    except Exception as e:
        print(f"Erro ao exibir detalhes dos cursos: {str(e)}")

def exibir_grafico_pizza_distribuicao():
    """
    Exibe um gráfico de pizza com a distribuição das categorias por carga horária.

    Returns:
        None
    """
    try:
        carga_horaria_por_categoria = df_categorizado.groupby('Categoria')['Carga Horária'].sum()

        labels = carga_horaria_por_categoria.index
        sizes = carga_horaria_por_categoria.values
        colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
        explode = (0.1, 0, 0, 0, 0)

        plt.figure(figsize=(8, 6))
        plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
        plt.axis('equal')
        plt.title('Distribuição das Categorias por Carga Horária')
        plt.show()
    except Exception as e:
        print(f"Erro ao exibir gráfico de pizza: {str(e)}")

def menu_interativo():
    """
    Menu interativo para análise de cursos por categoria ou exibição de gráfico.

    Returns:
        None
    """
    try:
        while True:
            print("\n### MENU INTERATIVO PARA ANÁLISE DE CURSOS ###")
            print("1. Tecnologia da Informação e Infraestrutura")
            print("2. Desenvolvimento e Programação")
            print("3. Ciência de Dados e Analytics")
            print("4. Outros")
            print("5. Exibir Gráfico de Distribuição por Categoria")
            print("6. Sair")

            opcao = input("Digite o número da opção desejada: ")

            if opcao == '1':
                exibir_detalhes_cursos("Tecnologia da Informação e Infraestrutura")
            elif opcao == '2':
                exibir_detalhes_cursos("Desenvolvimento e Programação")
            elif opcao == '3':
                exibir_detalhes_cursos("Ciência de Dados e Analytics")
            elif opcao == '4':
                exibir_detalhes_cursos("Outros")
            elif opcao == '5':
                exibir_grafico_pizza_distribuicao()
            elif opcao == '6':
                print("\nEncerrando o programa.")
                break
            else:
                print("\nOpção inválida. Por favor, tente novamente.")
    except Exception as e:
        print(f"Erro no menu interativo: {str(e)}")

# Montar o Google Drive
montar_drive()

# Definir o ID do arquivo CSV e o caminho onde será salvo
file_id = '1-8eHt9QhRMDesQz5rneQsSZePISex0Lj'  # Substitua pelo seu próprio ID
caminho_arquivo = '/content/dados_certificados.csv'

# Baixar o arquivo CSV do Google Drive
baixar_arquivo(file_id, caminho_arquivo)

# Carregar o arquivo CSV em um DataFrame
df = carregar_dataframe(caminho_arquivo)

# Categorizar os cursos do CSV
df_categorizado = categorizar_cursos(df)

# Executar o menu interativo
menu_interativo()
```
